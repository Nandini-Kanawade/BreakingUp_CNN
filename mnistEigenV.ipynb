```python
# Imports
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Load Mnist dataset
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
#Check Size
print(f"Training data shape: {x_train.shape}")
print(f"Training labels shape: {y_train.shape}")
print(f"Testing data shape: {x_test.shape}")
print(f"Testing labels shape: {y_test.shape}")
```

```python
#Concatenate the images and their labels
x = tf.concat([x_train, x_test], axis=0)
labels = tf.concat([y_train, y_test], axis=0)
image_data=x[0]
print(x[0])
```

```python
#in a dict store key-value label-image, take only 10 images of each set
digits = {}
for i in range(10):
    #print(i)
    digits[i] = x[labels == i][:10]

#Now we have a dictionary named digits 
    #key:0 value array all matrices with zero as label
    #key:1 value array all matrices with one as label
#digits.get(6) all the sixes
#digits.get(6)[0] fisrt image of 6
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml

# Load the MNIST dataset
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
X, y = mnist["data"], mnist["target"]

# Ensure y is a NumPy array of strings
y = np.array(y)

# Extract two images of digit "0"
zero_indices = np.where(y == '0')[0]
img1 = X[zero_indices[0]].reshape(28, 28)
img2 = X[zero_indices[1]].reshape(28, 28)

# Display the images
plt.subplot(1, 2, 1)
plt.title("First MNIST 0 Image")
plt.imshow(img1, cmap="gray")

plt.subplot(1, 2, 2)
plt.title("Second MNIST 0 Image")
plt.imshow(img2, cmap="gray")
plt.show()

# Compute the covariance matrix for each image
cov_matrix_img1 = np.cov(img1)
cov_matrix_img2 = np.cov(img2)

# Compute the eigenvalues and eigenvectors
eigenvalues_img1, eigenvectors_img1 = np.linalg.eig(cov_matrix_img1)
eigenvalues_img2, eigenvectors_img2 = np.linalg.eig(cov_matrix_img2)

# Filter out eigenvalues that are 0 and their corresponding eigenvectors
nonzero_indices_img1 = np.nonzero(eigenvalues_img1)
nonzero_indices_img2 = np.nonzero(eigenvalues_img2)

filtered_eigenvalues_img1 = eigenvalues_img1[nonzero_indices_img1]
filtered_eigenvectors_img1 = eigenvectors_img1[:, nonzero_indices_img1[0]]

filtered_eigenvalues_img2 = eigenvalues_img2[nonzero_indices_img2]
filtered_eigenvectors_img2 = eigenvectors_img2[:, nonzero_indices_img2[0]]

# Compare the filtered eigenvalues and eigenvectors
are_eigenvalues_similar = np.allclose(filtered_eigenvalues_img1, filtered_eigenvalues_img2)
are_eigenvectors_similar = np.allclose(filtered_eigenvectors_img1, filtered_eigenvectors_img2)

print(f"Are the eigenvalues similar? {are_eigenvalues_similar}")
print(f"Are the eigenvectors similar? {are_eigenvectors_similar}")
```
